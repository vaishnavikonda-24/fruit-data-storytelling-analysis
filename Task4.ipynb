```python
import pandas as pd

df = pd.read_csv('cleaned_fruit_dataset.csv')

df.head()
```

```python
print(type(df))
```

```python
print(df.shape)

print(df.info())

print(df.describe())
```

```python
total_fruits = len(df)

apple_count = len(df[df['Label']=="Apple"])

banana_count = len(df[df['Label']=="Banana"])

avg_weight = df['Weight'].mean()

print("Total Fruits =", total_fruits)

print("Apple Count =", apple_count)

print("Banana Count =", banana_count)

print("Average Weight =", avg_weight)
```

```python
import matplotlib.pyplot as plt

df['Label'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Fruit Distribution")

plt.show()
```

```python
plt.hist(df['Weight'])

plt.title("Weight Distribution")

plt.xlabel("Weight")

plt.ylabel("Frequency")

plt.show()
```

```python
df.groupby('Label')['Weight'].mean().plot(
    kind='bar'
)

plt.title("Average Weight by Fruit Type")

plt.ylabel("Weight")

plt.show()
```

```python
df['Fruit_Size'].value_counts().plot(
    kind='bar'
)

plt.title("Fruit Size Distribution")

plt.show()
```

```python
The objective of this project is to analyze fruit characteristics such as weight and size to identify patterns and gain insights into fruit classification.
```

```python
The dataset contains information about fruits including:

• Weight
• Colour
• Label
• Fruit Size
• Label Code

The dataset consists of Apple and Banana samples.
```

```python
Key Findings

• Total Fruits = 17

• Apples = 8

• Bananas = 9

• Average Weight = 334.06 g

• Maximum Weight = 382 g

• Minimum Weight = 277 g
```

```python
Business Insights

1. Bananas slightly outnumber Apples.

2. Weight is an important classification factor.

3. Data visualization simplifies understanding.

4. Dashboarding improves decision-making.
```

```python
from scipy.stats import ttest_ind

apple_weights = df[df['Label']=="Apple"]['Weight']

banana_weights = df[df['Label']=="Banana"]['Weight']

t_stat, p_value = ttest_ind(
    apple_weights,
    banana_weights
)

print("T Statistic =", t_stat)

print("P Value =", p_value)
```
